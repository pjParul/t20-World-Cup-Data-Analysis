# Batting Summary

```python
with open('Downloads/t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records=[]
    
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

**Create a column for OUT/NOT-OUT based on the values of Dismissal, if there is some value- out/ if it is blank - not out**

```python
df_batting['out/not-out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not-out")
df_batting.head()
```

```python
df_batting.drop(columns = ['dismissal'], inplace=True)
df_batting.head()
```

```python
df_batting.head(11)
```

```python
# remove the extra characters 'â€,\xa0' from batsman names
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',' '))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',' '))
df_batting.head(10)
```

**Join match_summary and batting_summary tables to map match_id from the 1st one**

```python
match_id_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']

match_id_dict
```

```python
# bring match_id column in batting_summary
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting.head()
```

```python
# export the table to CSV
df_batting.to_csv('fact_bating_summary.csv', index=False)
```

# Bowling Summary

```python
with open('Downloads/t20_json_files/t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
# map with match table, add new column - 'match_id'
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('fact_bowling.csv', index=False)
```
